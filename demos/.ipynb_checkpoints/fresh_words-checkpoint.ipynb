#Some more fun with crosswords

####Let's check out what new words have appeared in puzzles over the past year, and display those things over time

```python
from urllib2 import urlopen
from bs4 import BeautifulSoup as bs
import re
from datetime import date, datetime, timedelta
import pickle
import os
import sys
import time
from urllib import FancyURLopener
import urllib2
import json
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

scrape xwordinfo.com for new entries from every day in 2014

```python
# set date parameters

url_start = "http://www.xwordinfo.com/Fresh?date="

dates = []
start = date(2014,1,1); end = date(2014,12,31)
step = timedelta(days=1)
while start < end:
    dates.append(str(start.month) + "/" + str(start.day) + "/" + \
                 str(start.year))
    start += step

stats = []

# beautiful soup tinkering

for date in dates:
    
    url = url_start + date 
    ffdata = ["",""]

    sock = urlopen(url)
    lines = sock.read()
    sock.close()

    soup = bs(str(lines))
    b = soup.get_text()

    start = b.find("\n0"); end = b.find("\n1")
    words = b[start + 2 : end - 2].split()

    ffdata[0] = date
    ffdata[1] = words

    stats.append(ffdata)

# pickle saves us the trouble of reformatting everything

with open("ffstats_2014.txt", 'wb') as f:
    pickle.dump(stats, f)
f.close()
```

now i'd like some images of those entries

```python
# open up fresh words

with open("ffstats_2014.txt", 'rb') as f:
    info = pickle.load(f)
days = len(info) # 364

# for each day, choose new word to act as a search term

searchTerms = []; count= 0

for d in xrange(days):
    words = len(info[d][1])
    r = random.randint(0,words-1)
    searchTerm = info[d][1][r]
    searchTerms.append(searchTerm)

    # specify URL version, ajax specs

    class MyOpener(FancyURLopener): 
        version = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; it; rv:1.8.1.11) Gecko/20071127 Firefox/2.0.0.11'
    myopener = MyOpener()

    url = ('https://ajax.googleapis.com/ajax/services/search/images?' + \
           'v=1.0&q='+searchTerm+'&start='+"0"+'&userip=MyIP')
    request = urllib2.Request(url, None, {'Referer': 'testing'})
    response = urllib2.urlopen(request)

    # get results using JSON

    results = json.load(response)
    data = results['responseData']
    dataInfo = data['results']

    # get url and jpg for each word
    # actually had to rename files later due to initial zeroes

    for myUrl in dataInfo:
        count = count + 1
        # print myUrl['unescapedUrl']
        myopener.retrieve(myUrl['unescapedUrl'],str(count)+'.jpg')
        break
```

```python

with open("terms.txt", 'wb') as f:
    pickle.dump(searchTerms, f)
f.close()
```

plot a series of images

```python
# get a list of filenames

picfiles = []
for top, dirs, files in os.walk('.'):
    for nm in files:
        picfiles.append(nm)
```

```python

def draw(i):
    im = plt.imread(picfiles[i])
    img = plt.imshow(im)
    plt.title(searchTerms[i])
    plt.draw()

for i in range(5):
    draw(i)
    plt.show()
```
